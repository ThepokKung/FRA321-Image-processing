# **Import Libary**

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

# **Load Image**

```python
ori_img = cv2.imread('image/inside-the-box.jpg',cv2.IMREAD_COLOR)
ori_img2 = cv2.imread('image/inside-the-box.jpg',0)


alpha = 6.0  # Contrast control (1.0-3.0)
beta = 75    # Brightness control (0-100)

bright_image = cv2.convertScaleAbs(ori_img, alpha=alpha, beta=beta)
cv2.imwrite("image/brightened_image.jpg", bright_image)

denoised = cv2.medianBlur(bright_image, 3)
cv2.imwrite("image/denoised_image.jpg", denoised)

equalizeHist = cv2.equalizeHist(ori_img2)
cv2.imwrite("image/equalizeHist_image.jpg", equalizeHist)

plt.figure(figsize=(20, 5))
plt.subplot(1, 4, 1)
plt.imshow(ori_img)
plt.axis('off')
plt.title('inside-the-box (real)')

plt.subplot(1, 4, 2)
plt.imshow(bright_image)
plt.axis('off')
plt.title('inside-the-box (bright)')

plt.subplot(1, 4, 3)
plt.imshow(denoised)
plt.axis('off')
plt.title('inside-the-box (denoised)')

plt.subplot(1, 4, 4)
plt.imshow(equalizeHist)
plt.axis('off')
plt.title('inside-the-box (equalizeHist)')

plt.show()
```
