# **Import Libary**

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

# **Unsharp Masking & Highboot**

```python
ori_img = cv2.imread('image/mega_space_molly.jpg',0)

blurred = cv2.GaussianBlur(ori_img, (5, 5), 1.0)

mask = cv2.subtract(ori_img,blurred)

ku = 1  #  1  Unsharp
unsharp_image = cv2.addWeighted(ori_img, 1.0, mask, ku, 0)

kh = 4.5  #  1++  Highboot
hightboot_image = cv2.addWeighted(ori_img, 1.0, mask, kh, 0)

plt.figure(figsize=(20, 5))
plt.subplot(1, 5, 1)
plt.imshow(ori_img,cmap='gray')
plt.axis('off')
plt.title('Original Image')

plt.subplot(1, 5, 2)
plt.imshow(blurred,cmap='gray')
plt.axis('off')
plt.title('blurred Image')

plt.subplot(1, 5, 3)
plt.imshow(mask,cmap='gray')
plt.axis('off')
plt.title('mask Image')

plt.subplot(1, 5, 4)
plt.imshow(unsharp_image,cmap='gray')
plt.axis('off')
plt.title('unsharp Image (k = 1)')

plt.subplot(1, 5, 5)
plt.imshow(hightboot_image,cmap='gray')
plt.axis('off')
plt.title('hightboot Image (k = 4.5)')

cv2.imwrite("image/gray.jpg",ori_img)
cv2.imwrite("image/blurred.jpg", blurred)
cv2.imwrite("image/mask.jpg", mask)
cv2.imwrite("image/unsharp.jpg",unsharp_image)
cv2.imwrite("image/hightboot.jpg",hightboot_image)
```
